```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import pearsonr
```

```python
train=pd.read_csv('../cv2/train1x.csv')
y=pd.read_csv('../input/train.csv')['fault_severity'].values
```

```python
cmap={}
for i in train.columns.values:
    cmap[i]=abs(pearsonr(train[i],y)[0])
```

```python
newfea=[]
fea=[i for i in train.columns.values]
for c,i in enumerate(fea[:-1]):
    for d,j in enumerate(fea[c+1:]):
        cc=abs(pearsonr(train[i]-train[j],y)[0])
        if cc>cmap[i]*1.2 and cc>cmap[j]*1.2 and cc>0.1:
            train['%s_%s'%(i,j)]=train[i]-train[j]
            print c,c+d,i,j,cc,cmap[i],cmap[j]
            newfea.append('%s_%s'%(i,j))
            break
```

```python
train[newfea].to_csv('train2.csv',index=False)
```
